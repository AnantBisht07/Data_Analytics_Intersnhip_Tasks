```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.plot('SepalLengthCm','Species',data=df)
plt.show()
```

```python
df.groupby('Species').size().plot(kind='bar',
                                 figsize = (6,6),
                                 title = 'TOTAL SPECIES',
                                 color='blue')
plt.show()
```

###### This Bar graph showing the sum ofSepalLength for each species in the dataframe.

```python
df.groupby('Species').sum()['SepalLengthCm'].plot(kind='bar',
                                                 figsize=(5,5),
                                                 color='red')
plt.show()
```

##### As We can see that the mean SepalLengthCm is highest for the virginica species, followed by the versicolor species, and lowest for the setosa species. 

```python
df.groupby('Species').mean()['SepalLengthCm'].plot(kind='bar',
                                                 figsize=(5,5),
                                                  color='blue')
plt.show()
```

```python
df.groupby('Id').mean()['SepalLengthCm'].plot(kind='bar',
                                                  color='blue')
plt.show()
```

###### This bar graph showing the Total size of the Id datatframe.

```python
df.groupby('Id').size().plot(kind='bar',
                                 figsize = (6,6),
                                 title = 'TOTAL Ids',
                                 color='yellow')
plt.show()
```
